```python
import requests

def verify_with_ban(identifiant, address):
    url = f"https://api-adresse.data.gouv.fr/search/?q={address}"
    response = requests.get(url).json()
    if response['features']:
        ban_id = response['features'][0]['properties']['id']
        return identifiant == ban_id
    return False
```

```python
a['verified'] = a.apply(lambda row: verify_with_ban(row['identifiant_ban'], row['adresse_ban']), axis=1)
```

```python
import requests

def verify_with_ban(address):
    url = f"https://api-adresse.data.gouv.fr/search/?q={address}&limit=1"
    response = requests.get(url).json()
    # return response
    if response['features']:
        ban_id, ban_add = response['features'][0]['properties']['id'],response['features'][0]['properties']['label']
        return ban_id, ban_add
    return None, None
```
