```python
import requests

def verify_with_ban(identifiant, address):
    url = f"https://api-adresse.data.gouv.fr/search/?q={address}"
    response = requests.get(url).json()
    if response['features']:
        ban_id = response['features'][0]['properties']['id']
        return identifiant == ban_id
    return False
```

```python
import unicodedata
import re

def normaliser_adresse(adresse):
    if not adresse:
        return ""

    adresse = adresse.upper()
    adresse = unicodedata.normalize('NFKD', adresse)
    adresse = adresse.encode('ASCII', 'ignore').decode('utf-8')

    adresse = adresse.replace("'", " ")
    adresse = re.sub(r'[^A-Z0-9 -]', '', adresse)

    adresse = " ".join(adresse.split())

    return adresse
```

```python
df_test['adresse_normalisee'] = df_test['add_ban'].apply(normaliser_adresse)
```
